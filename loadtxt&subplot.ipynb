```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.zeros([10, 8]);
A[:, 0] = np.linspace(0, 9, 10);
A[:, 2] = 2 * np.linspace(0, 9, 10);
A[:, 4] = 1.5 * np.linspace(0, 9, 10);
A[:, 6] = 10 * np.linspace(0, 9, 10);
A[:, 1] = np.random.rand(10);
A[:, 3] = np.random.rand(10);
A[:, 5] = np.random.rand(10);
A[:, 7] = np.random.rand(10);

np.savetxt('points.txt', A)
del A
```

```python
B = np.loadtxt('points.txt')
print(B)
```

```python
fig = plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.plot(B[:,0], B[:,1])
plt.title('subplot1: cols 1 & 2')
plt.xlabel('depth')
plt.ylabel('heigth')

plt.subplot(2, 2, 2)
plt.plot(B[:,2], B[:,3])
plt.title('cols 3 & 4')
plt.xlabel('depth2')
plt.ylabel('heigth2')

plt.subplot(2, 2, 3)
plt.plot(B[:,4], B[:,5], label='laleyendaquequiero')
plt.title('cols 5 & 6')
plt.xlabel('depth3')
plt.ylabel('heigth3')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(B[:,6], B[:,7])
plt.title('cols 7 & 8')
plt.xlabel('depth4')
plt.ylabel('heigth4')
# plt.show()
```

```python
fig = plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.plot(B[:,0], B[:,1], label='laleyendaquequiero')
plt.title('$subplot1: cols 1 & 2$')
plt.xlabel('$depth$')
plt.ylabel('$heigth$')
plt.legend()

plt.subplot(1, 4, 2)
plt.plot(B[:,2], B[:,3], 'r',  label='$laleyendaenalatex$')
plt.title('$cols 3\ &\ 4$')
plt.xlabel('$depth^{2}$')
plt.ylabel('$heigth^{2}$')
plt.legend()

plt.subplot(1, 4, 3)
plt.plot(B[:,4], B[:,5], 'g*')
plt.title('$cols 5\ &\ 6$')
plt.xlabel('$depth3$')
plt.ylabel('$heigth3$')

plt.subplot(1, 4, 4)
plt.plot(B[:,6], B[:,7], 'y-^', lw='2')
plt.title('$cols 7\ &\ 8$')
plt.xlabel('$depth_{4}$')
plt.ylabel('$heigth_{4}$')
# plt.show()
```
